```python
import networkx as nx
graphSample = nx.DiGraph()

#Node declaration
graphSample.add_nodes_from([
  (0, {"color": "lightgrey", "size": 400}),
  (1, {"color": "lightgrey", "size": 400}),
  (2, {"color": "lightgrey", "size": 400}),
  (3, {"color": "lightgrey", "size": 400}),
  (4, {"color": "lightgrey", "size": 400}),
  (5, {"color": "lightgrey", "size": 400}),
  (6, {"color": "lightgrey", "size": 400})
])

#Edges
graphSample.add_edges_from([
  (0, 1),
  (1, 2),
  (1, 0),
  (1, 3),
  (2, 4),
  (3, 0),
  (4, 4),
  (4, 3),
  (5, 6),
  (2, 6),
  (6, 0)
])

node_colors = nx.get_node_attributes(graphSample, "color").values()
colors = list(node_colors)
node_sizes = nx.get_node_attributes(graphSample, "size").values()
sizes = list(node_sizes)

#Plotting Graph
nx.draw(graphSample, with_labels=True, node_color=colors, node_size=sizes)
```
